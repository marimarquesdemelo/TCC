```python
# importando bibliotecas
import pandas as pd
import glob
from pysus.ftp.databases.sinan import SINAN
```

```python
# carregando dados do DATASUS
sinan = SINAN().load()
```

```python
# obtendo os arquivos de sífilis congênita de 2013 a 2022
sinan.get_files(dis_code=["SIFC"], year=[2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
```

```python
files_sifc = sinan.get_files(dis_code="SIFC", year=[2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
sifc_2013, sifc_2014, sifc_2015, sifc_2016, sifc_2017, sifc_2018, sifc_2019, sifc_2020, sifc_2021, sifc_2022 = files_sifc
```

```python
# realizando o downloas dos arquivos
sinan.download(files_sifc)
```

```python
# concatenando os arquivos parquet
arquivos = glob.glob("/root/pysus/SIFC*.parquet")
# 'arquivos' agora é um array com o nome de todos os .parquet que se iniciam com SIFC existentes no diretório indicado'
df = pd.concat(map(pd.read_parquet, arquivos))
```

```python
# ordenando os registros por ano
df = df.sort_values(by=['NU_ANO'])
```
